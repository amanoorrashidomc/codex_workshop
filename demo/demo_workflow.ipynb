## Step 1: User-editable inputs
Edit these values for your real workflow.

````python
from pathlib import Path

# USER EDIT: set your PBIX file path
pbix_path = Path('../input/report.pbix')
# USER EDIT: text to find
find_text = 'OldValue'
# USER EDIT: replacement text
replace_text = 'NewValue'

mock_json = Path('sample_data/extracted_mock/example.json')
mock_json
````

## Step 2: Simulated workflow (when real PBIX is unavailable)
This uses a sample extracted JSON file to demonstrate find/replace behavior.

````python
text = mock_json.read_text(encoding='utf-8')
count = text.count(find_text)
updated = text.replace(find_text, replace_text)
print(f'[INFO] JSON files scanned: 1')
print(f'[INFO] JSON files changed: {1 if count > 0 else 0}')
print(f'[INFO] Total replacements: {count}')
print('--- updated preview ---')
print(updated)
````
